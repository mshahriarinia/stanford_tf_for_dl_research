## Convolution Layer

__Convolve__ the filter with the image i.e. “slide over the image spatially, computing dot products”
- Filters always extend the full depth of the input volume (same # of channels): For a 2-D image of size h x w which has k channels (e.g. 3 colors for RGB), we slide c x c filter with the same number of channels (e.g. 32x32x__3__ and 5x5x__3__). We refer to filter as *W*
- __1 number:__ the result of taking a dot product between the filter and a small 5x5x3 chunk of the image (i.e. 5*5*3 = 75-dimensional dot product + __bias__): *W^T X + b*
- convolve (slide) over all spatial locations. Since the reult of sliding the filter at each location is just one number; This will result in 28x28x1. The result is called *activation map*. (loses 2 pixels from each side since the filter is 5x5 and from its center there are two pixels on left and two pixels on right)
- Consider a second filter and convolve (slide) over all spatial locations. So we would have two activation maps. 
- For example, if we had 6 5x5 filters (k), we’ll get 6 separate activation maps. We stack these up to get a “new image” of size 28x28x6!

__Pooling Layer__: Since by applying k filters we get k activation maps of roughly the same size as of the image we down sample them to make them computationally manageable. It's like reducing the size of an image by resizing it in half. 
- makes the representations smaller and more manageable
- operates over each activation map independently

For example in a 224x224x64 activation maps, by pooling (__down sampling__) (e.g. 2x2) __non-overlapping__ we get 112x112x64 activation maps. 
- Each activation map holds and abstract image representing and *firing* on certain characteristics of the dataset

__Max Pooling__ To downsample an activation map, pick the maximum value of a pixel and its surroundings
